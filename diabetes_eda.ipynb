## Invalid data processing

```python
df.head()
```

```python
print(df[df.Tricep == 0].shape[0])
print(df[df.Tricep == 0].index.tolist())
```

```python
opt_data = df[(df.BloodPressure != 0) & (df.BMI != 0) & (df.Glucose != 0)]
opt_data.shape
```

```python
df['Glucose'] = df['Glucose'].apply(lambda x: np.nan if x == 0 else x)
df['BloodPressure'] = df['BloodPressure'].apply(lambda x: np.nan if x == 0 else x)
df['Tricep'] = df['Tricep'].apply(lambda x: np.nan if x == 0 else x)
df['Insulin'] = df['Insulin'].apply(lambda x: np.nan if x == 0 else x)
df['BMI'] = df['BMI'].apply(lambda x: np.nan if x == 0 else x)
```

```python
df.isnull().sum()
```

```python
df_fillna = df['Tricep'].fillna('No data')
df_fillna
```

```python
df.fillna('No data', inplace=True)
df.head()
```

```python
df_fillna = df.fillna(np.mean(df))
df_fillna.head()
```

```python
df_dropna = df.dropna()
df_dropna.head()
```
